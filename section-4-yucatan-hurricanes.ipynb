```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("https://portal.edirepository.org/nis/dataviewer?packageid=knb-lter-hfr.71.23&entityid=ab0fe2bf4f3ad850371ccb9c69d78469")
```

```python
df.head()
```

```python
(df.loc[df['ss'] == 5]
        .assign(duration=pd.to_datetime(df['end.date']) - pd.to_datetime(df['start.date']))
       )[['name', 'ss', 'duration']]
```

```python
plt.figure()
plt.scatter(pd.to_datetime(df['start.date']), df['ss'])


plt.title('Start Date vs SS')
plt.xlabel('Start Date')
plt.ylabel('Saffir-Simpson Scale')


plt.xticks(rotation=45)
plt.yticks([1, 2, 3, 4, 5])

plt.show()
```
